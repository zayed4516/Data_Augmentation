## Data Augmentation

**Definition:**  
Data augmentation is a technique for artificially increasing the training set by creating modified copies of existing data.  
It can involve minor changes to the dataset or use deep learning techniques to generate new data points.

---

### Augmented vs. Synthetic Data

| Aspect              | Augmented Data | Synthetic Data |
|---------------------|----------------|----------------|
| **Definition**      | Modified versions of existing data to increase diversity. | Artificially generated data without direct reliance on original samples. |
| **Example (Images)**| Rotations, flips, color adjustments on existing images. | Fully generated images using simulations or generative models. |
| **Purpose**         | Helps models generalize better by exposing them to varied but related data. | Allows testing and improving algorithms without risking privacy or security. |

---

**Note:**  
Data augmentation is not limited to images â€” it can be applied to **audio**, **video**, **text**, and other data types.



<p style="background-color:#348ceb; color:white; padding:10px;">
  <b style="font-size: 30px;"> Why is Data Augmentation Important?</b>
</p>



Data augmentation helps machine learning models perform better by making the most of existing data. It:  
- **Prevents overfitting** by exposing the model to more varied data.  
- **Improves accuracy** and robustness in real-world scenarios.  
- **Creates diversity** in training data, especially when datasets are small or imbalanced.  
- **Simulates real-world variations** without the cost of collecting new data.  
- Enables models to **generalize more effectively**.


<p style="background-color:#3464eb; color:white; padding:10px;">
  <b style="font-size: 30px;"> When Should You Use Data Augmentation?</b>
</p>



You should consider data augmentation when:  
- The model is **overfitting** to the training data.  
- The **initial training set is too small** to capture all variations.  
- You want to **improve the model's accuracy** and robustness.  
- You need to **reduce operational costs** for labeling and cleaning large datasets.


<p style="background-color:#4034eb; color:white; padding:10px;">
  <b style="font-size: 30px;">  Limitations of Data Augmentation</b>
</p>



1. **Bias Persistence**  
   The biases in the original dataset persist in the augmented data. If the source data lacks diversity or contains skewed distributions, augmentation will replicate these biases instead of removing them.

2. **High Quality Assurance Cost**  
   Quality assurance for data augmentation can be expensive, as it often requires manual inspection or automated validation to ensure that the augmented samples remain meaningful and correctly labeled.

3. **Research and Development Effort**  
   Developing advanced augmentation systems requires significant R&D. For example, generating high-resolution images using GANs can be technically challenging and resource-intensive.

4. **Difficulty in Finding Effective Approaches**  
   Finding the right augmentation strategy for a specific problem can be challenging. The wrong approach may harm model performance instead of improving it.

<p style="background-color:#7434eb; color:white; padding:10px;">
  <b style="font-size: 30px;">  Data Augmentation Techniques</b>
</p>

-------------------------------------------------------------------------------


<p style="background-color:#b734eb; color:white; padding:10px;">
  <b style="font-size: 30px;">  Audio Data Augmentation</b>
</p>


1. **Noise Injection**  
   Add Gaussian or random noise to the audio dataset to improve model robustness against background sounds.

2. **Shifting**  
   Shift the audio signal to the left (fast forward) or right by a random number of seconds.

3. **Changing the Speed**  
   Stretch or compress the audio time series by a fixed rate, making it play faster or slower.

4. **Changing the Pitch**  
   Randomly change the pitch of the audio without altering its duration, simulating variations in speaker tone or instrument tuning.



<p style="background-color:#b734eb; color:white; padding:10px;">
  <b style="font-size: 30px;">  Text Data Augmentation</b>
</p>



1. **Word or Sentence Shuffling**  
   Randomly change the position of words or sentences to create new variations of the text.

2. **Word Replacement**  
   Replace words with their synonyms to preserve meaning while increasing data diversity.

3. **Syntax-tree Manipulation**  
   Paraphrase the sentence using the same words but with a different syntactic structure.

4. **Random Word Insertion**  
   Insert additional words at random positions in the text.

5. **Random Word Deletion**  
   Delete words at random to create shortened or incomplete sentences, helping the model handle missing data.



<p style="background-color:#b734eb; color:white; padding:10px;">
  <b style="font-size: 30px;">  Image Augmentation</b>
</p>

1. **Geometric Transformations**  
   Randomly flip, crop, rotate, stretch, and zoom images.  
   *Note:* Be cautious when applying multiple transformations on the same image, as this can sometimes reduce model performance.

2. **Color Space Transformations**  
   Randomly change RGB color channels, adjust contrast, and modify brightness.

3. **Kernel Filters**  
   Randomly change image sharpness or apply blurring to simulate different focus levels.

4. **Random Erasing**  
   Delete or mask out some part of the initial image to improve robustness.

5. **Mixing Images**  
   Blend or mix multiple images together to create new composite samples.


<p style="background-color:#b734eb; color:white; padding:10px;">
  <b style="font-size: 30px;">  Advanced Data Augmentation Techniques</b>
</p>



1. **Generative Adversarial Networks (GANs)**  
   Used to generate entirely new data points or images.  
   - Can create synthetic data without directly requiring existing samples for transformation.  
   - Useful for expanding datasets, especially in domains where collecting real data is difficult.

2. **Neural Style Transfer**  
   Uses a series of convolutional layers to deconstruct images and separate **content** from **style**.  
   - Allows applying the style of one image (e.g., an artwork) to the content of another image.  
   - Commonly used in creative applications and data augmentation for visual diversity.

  <b style="font-size: 30px;">Data Augmentation Applications</b>
</p>



Data augmentation can be applied to all machine learning applications where acquiring quality data is challenging.  
It also helps improve model robustness and performance across different fields of study.

---

### 1. Healthcare

Acquiring and labeling medical imaging datasets is often **time-consuming** and **expensive**.  
Additionally, you need a subject matter expert (SME) to validate the dataset before performing data analysis.  

Using **geometric** and **other transformations** can help you train robust and accurate machine learning models.

**Example: Pneumonia Classification**  
- **Useful transformations:** random cropping, zooming, stretching, and color space transformations to improve model performance.  
- **Caution:** Certain augmentations can harm performance. For example:  
  - Random rotation  
  - Reflection along the x-axis  
  These are **not recommended** for X-ray imaging datasets as they can distort anatomical orientation.

### 2. Self-Driving Cars

There is limited real-world data available for self-driving cars due to safety, privacy, and cost constraints.  
Companies use **simulated environments** to generate **synthetic data** with techniques such as **reinforcement learning**.

Benefits:  
- Allows training and testing of machine learning models without risking human safety.  
- Provides controlled scenarios to test edge cases (e.g., rare weather conditions, sudden obstacles).  
- Helps address **data security** and **privacy** concerns.

Example:  
Simulations can generate varied road, traffic, and weather conditions for robust autonomous driving model training.

### 3. Natural Language Processing (NLP)

Text data augmentation is commonly used when:  
- There is **limited quality data** available.  
- Improving performance metrics is a priority.  
- Working with **low-resource languages** that lack large annotated datasets.

**Common techniques:**  
- **Synonym augmentation:** Replace words with their synonyms.  
- **Word embedding augmentation:** Use vector similarity to find and replace words with semantically similar ones.  
- **Character swap:** Randomly swap adjacent characters in words.  
- **Random insertion:** Insert random words at various positions.  
- **Random deletion:** Remove words at random to simulate missing data.

**Benefit:**  
These techniques help models generalize better and improve performance even with small or imbalanced text datasets.

### 4. Automatic Speech Recognition (ASR)

In sound classification and speech recognition, **data augmentation** can significantly improve model performance â€” even for **low-resource languages**.

**Common techniques:**  
- **Noise injection:** Add background noise to simulate real-world environments.  
- **Time shifting:** Shift the audio forward or backward by a small amount.  
- **Speed and pitch changes:** Vary speaking speed or pitch without distorting the audio too much.  
- **SpecAugment:** Apply time and frequency masking directly to spectrograms.

**Benefit:**  
Makes ASR systems more robust to different accents, recording conditions, and background environments.

<p style="background-color:#f01f26; color:white; padding:10px;">
  <b style="font-size: 30px;">Ethical Implications of Data Augmentation</b>
</p>



While data augmentation is powerful, it also raises important ethical issues:

1. **Bias Amplification**  
   If the original data is biased, augmentation can make the problem worse, leading to unfair results for certain groups.

2. **Privacy Risks**  
   Synthetic data might still contain sensitive details from the original dataset, risking privacy violations.

3. **Data Authenticity**  
   Poorly applied augmentations can create unrealistic or misleading data, reducing model reliability.

4. **Transparency**  
   Clearly document what augmentation techniques were used so stakeholders know how the data was changed or generated.

5. **Fairness and Equity**  
   Augmented data should represent diverse populations and scenarios to avoid unfairness.

6. **Regulatory Compliance**  
   Ensure augmented datasets follow data protection laws and ethical guidelines to prevent legal issues.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
dataset_url="https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir=tf.keras.utils.get_file('flower_photos',origin=dataset_url,cache_dir='.',untar=True)
```

```python
import pathlib

data_dir=pathlib.Path(data_dir)/ "flower_photos"
```

```python
roses=list(data_dir.glob('roses/*'))
roses[:5]
```

```python
tulips=list(data_dir.glob('tulips/*'))
PIL.Image.open(str(tulips[0]))
```

```python
flower_images_dict={
    'roses':list(data_dir.glob('roses/*')),
    'daisy':list(data_dir.glob('daisy/*')),
    'dandelion':list(data_dir.glob('dandelion/*')),
    'sunflowers':list(data_dir.glob('sunflowers/*')),
    'tulips':list(data_dir.glob('tulips/*'))
}
```

```python
flowers_labels_dict={
    'roses':0,
    'daisy':1,
    'dandelion':2,
    'sunflowers':3,
    'tulips':4
}

```

```python
img=cv2.imread(str(flower_images_dict['roses'][0]))
img
```

```python
img.shape
```

```python
cv2.resize(img,(180,180)).shape
```

```python
for flower_name,images in flower_images_dict.items():
  print(flower_name)
  print(len(images))
  print('*'*30)
```

```python
X,y=[],[]
h,w=180,180
for flower_name,images in flower_images_dict.items():
  for img in images:
    img=cv2.imread(img)
    resized_img=cv2.resize(img,(h,w))
    X.append(resized_img)
    y.append(flowers_labels_dict[flower_name])
```

```python
y[:4]
```

```python
X[0]
```

```python
X=np.array(X)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train_scaled=X_train/255
X_test_scaled=X_test/255
```

```python
X_train_scaled[0]
```

```python
model=Sequential([
    layers.Conv2D(16,3,padding='same',activation='relu'),
    layers.MaxPool2D(),
    layers.Conv2D(32,3,padding='same',activation='relu'),
    layers.MaxPool2D(),
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPool2D(),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(5)
])
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model.fit(X_train_scaled,y_train,epochs=15)
```

## find the model cause overfitting

```python
model.evaluate(X_test_scaled,y_test)
```

```python
predictions=model.predict(X_test_scaled)
predictions
```

## build model with augmentation

```python
img_height=180
img_width=180

data_augmentation_all = keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal", input_shape=(img_height, img_width, 3)),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1)
])
```

## We prevented overfitting, and the accuracy can be increased by gradually increasing the number of epochs.


```python
model=Sequential([
    data_augmentation_all,
    layers.Conv2D(16,3,padding='same',activation='relu'),
    layers.MaxPool2D(),
    layers.Conv2D(32,3,padding='same',activation='relu'),
    layers.MaxPool2D(),
    layers.Conv2D(64,3,padding='same',activation='relu'),
    layers.MaxPool2D(),
    layers.Dropout(0.2),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(5)
])

model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])

model.fit(X_train_scaled,y_train,epochs=15)
```

We noticed that the accuracy in the evaluation increased from **0.5893** to **0.7004**, which indicates that we solved the overfitting problem.


```python
model.evaluate(X_test_scaled,y_test)
```

```python
predictions=model.predict(X_test_scaled)
predictions
```
